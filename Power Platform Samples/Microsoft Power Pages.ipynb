# Power Pages
***

Microsoft Power Pages is a robust, enterprise-level, low-code platform delivered as a software-as-a-service (SaaS), designed for the development, hosting, and management of contemporary business websites intended for external audiences. Whether you identify as a low-code enthusiast or a seasoned developer, Power Pages enables you to swiftly design, customize, and deploy websites that seamlessly function across various web browsers and devices. [Register](https://nonprofit.microsoft.com/en-us/getting-started) to claim your annual subscription grants for Non-Profits or [sign up for free trial](https://learn.microsoft.com/en-us/power-pages/getting-started/trial-signup).

## Power Pages Capabilities
___

#### 1. Simplified authoring experience for make

Quickly create new sites directly from the Power Pages home page by using the default template, or choose from existing industry-based starter templates. Learn more on how to [create a site](https://learn.microsoft.com/en-us/power-pages/getting-started/create-manage)

#### 2. Design studio
Build powerful and engaging sites without writing a single line of code.

#### 3. Responsive rendering
Power Pages is based on Bootstrap, which natively provides support for building websites that are responsive, mobile-friendly, and available in various form factors.

#### 4. Security and governance
Power Pages' inbuilt security is its core. It allows organizations to securely enable access of their business data to their users (internal or external) through Power Pages authorization rules. More information: [Power Pages security](https://learn.microsoft.com/en-us/power-pages/security/power-pages-security)


#### 5. Integration with other Microsoft Power Platform components
   Power Pages provides deep integration with other [Microsoft Power Platform](https://learn.microsoft.com/en-us/power-platform/) components.





## Create a site with Power Pages
---

1. Go to [Power Pages](https://make.powerpages.microsoft.com/environments/f0dabaed-ac54-e926-8d2e-6f6a1f099619/portals/home)

Note:The first time you visit the site, select the [Try it for free button to sign up for a free, 30-day trial](https://learn.microsoft.com/en-us/power-pages/getting-started/trial-signup).

2. Select the [Microsoft Dataverse environment](https://learn.microsoft.com/en-us/power-platform/admin/environments-overview) where you'd like to create a site. 

3. On the home page, select + Create a site. 

Note: You can choose to create site from available templates, start from blank or Copilot.



4. Review the available templates. For more details about each template, hover over the template and choose Preview template. Follow the different views across devices to preview the template experience.

**If none of the business need templates match what you're looking for, choose one of the Starter layout templates with cross-industry solutions, or choose Blank page to customize the website from scratch.** More information: [Power Pages templates](https://learn.microsoft.com/en-us/power-pages/templates/)

5. After you've found the best template for your business needs, select Choose this template.

6. Validate the default site name and web address, then select Done.





**Note: It might take a few moments for your new site to be provisioned. You'll be able to modify the name and web address later.**

7. After the site is created, you can begin to edit or preview your site.

### Portal Management

You can use the Portal Management app, a model-driven Power App, for advanced configuration of your Power Pages sites. When you create a site with a new trial environment, the site metadata for all the templates is preloaded. It appears as website records in the [Portal Management app](https://learn.microsoft.com/en-us/power-pages/configure/portal-management-app).
